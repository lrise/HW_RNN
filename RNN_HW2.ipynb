```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import psutil
import os

# For text preprocessing
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# For model building
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, GRU, Dense, Dropout, SpatialDropout1D, Bidirectional, GlobalMaxPooling1D
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from sklearn.utils.class_weight import compute_class_weight

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Custom callback for F1 score monitoring
class F1Metrics(tf.keras.callbacks.Callback):
    def __init__(self, validation_data):
        super(F1Metrics, self).__init__()
        self.validation_data = validation_data

    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        val_pred = (self.model.predict(self.validation_data[0]) > 0.5).astype(int)
        val_f1 = f1_score(self.validation_data[1], val_pred)
        logs['val_f1'] = val_f1
        print(f" — val_f1: {val_f1:.4f}")

# 1. Load and explore the data
print("Loading data...")
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

# Display first few rows
print("\nFirst few rows of training data:")
train_df.head()

# Check for missing values
print("\nMissing values in train data:")
print(train_df.isnull().sum())

# Data distributions
print("\nTarget distribution in training data:")
print(train_df['target'].value_counts())
print(train_df['target'].value_counts(normalize=True) * 100)

# Calculate class weights
class_weights = compute_class_weight('balanced', classes=np.unique(train_df['target']), y=train_df['target'])
class_weight_dict = dict(zip(np.unique(train_df['target']), class_weights))
print("\nClass weights:", class_weight_dict)

# 2. Feature Engineering - Add additional features that might help with classification
def extract_additional_features(df):
    # Create new features that might help disaster detection
    features = pd.DataFrame()

    # Text length might be indicative (urgent messages tend to be concise)
    features['text_length'] = df['text'].apply(lambda x: len(x) if isinstance(x, str) else 0)

    # Count of special emergency-related characters
    features['exclamation_count'] = df['text'].apply(lambda x: x.count('!') if isinstance(x, str) else 0)
    features['question_count'] = df['text'].apply(lambda x: x.count('?') if isinstance(x, str) else 0)
    features['hashtag_count'] = df['text'].apply(lambda x: x.count('#') if isinstance(x, str) else 0)

    # Count of emergency-related keywords (customize this list)
    emergency_keywords = ['emergency', 'disaster', 'fire', 'flood', 'earthquake', 'evacuation',
                         'danger', 'warning', 'injured', 'casualties', 'damage', 'destroyed']

    features['emergency_word_count'] = df['text'].apply(
        lambda x: sum(1 for word in str(x).lower().split() if word in emergency_keywords)
    )

    return features

# Extract additional features
print("Extracting additional features...")
train_features = extract_additional_features(train_df)
test_features = extract_additional_features(test_df)

# Display the additional features
print("\nAdditional features (first few rows):")
print(train_features.head())

# 3. Improved Text Preprocessing
def improved_preprocess_text(text):
    if isinstance(text, str):
        # Convert to lowercase
        text = text.lower()

        # Remove URLs
        text = re.sub(r'https?://\S+|www\.\S+', '', text)

        # Remove HTML tags
        text = re.sub(r'<.*?>', '', text)

        # Keep important punctuation that may signal distress or urgency
        # Only remove some punctuation, keep: ! ? # @ $ % -
        punctuation_to_remove = string.punctuation.replace('!', '').replace('?', '').replace('#', '')
        text = text.translate(str.maketrans('', '', punctuation_to_remove))

        # Remove mentions but keep hashtag content
        text = re.sub(r'@\w+', '', text)
        text = re.sub(r'#(\w+)', r'\1', text)

        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text).strip()

        # Tokenize
        tokens = word_tokenize(text)

        # Selective stopword removal - keep important ones like 'not', 'no', 'very'
        important_stopwords = {'not', 'no', 'very', 'against', 'down', 'most', 'few'}
        stop_words = set(stopwords.words('english')) - important_stopwords
        tokens = [word for word in tokens if word not in stop_words]

        # Apply lemmatization more selectively - only to certain parts of speech
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens]

        # Rejoin tokens
        text = ' '.join(tokens)

        return text
    return ""

# Apply improved preprocessing to the text columns
print("Preprocessing text data...")
train_df['cleaned_text'] = train_df['text'].apply(improved_preprocess_text)
test_df['cleaned_text'] = test_df['text'].apply(improved_preprocess_text)

# Display some examples of cleaned text
print("\nExamples of cleaned text:")
for i in range(5):
    print(f"Original: {train_df['text'].iloc[i]}")
    print(f"Cleaned: {train_df['cleaned_text'].iloc[i]}")
    print(f"Target: {train_df['target'].iloc[i]}")
    print("---")

# 4. Tokenization and Sequence Generation
# Tokenize the text with increased vocab size
MAX_VOCAB_SIZE = 15000  # Increased from 10000
MAX_SEQUENCE_LENGTH = 100

tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE, oov_token='<OOV>')
tokenizer.fit_on_texts(train_df['cleaned_text'])

# Convert text to sequences
train_sequences = tokenizer.texts_to_sequences(train_df['cleaned_text'])
test_sequences = tokenizer.texts_to_sequences(test_df['cleaned_text'])

# Pad sequences
X_train_text = pad_sequences(train_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')
X_test_text = pad_sequences(test_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')

# Combine with additional features
X_train = np.hstack((X_train_text, train_features.values))
X_test = np.hstack((X_test_text, test_features.values))

# Target variable
y_train = train_df['target'].values

# Get vocabulary size
vocab_size = min(MAX_VOCAB_SIZE, len(tokenizer.word_index) + 1)
print(f"Vocabulary size: {vocab_size}")

# Split the data into training and validation sets
X_train_data, X_val_data, y_train_data, y_val_data = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
)

print(f"Training data shape: {X_train_data.shape}")
print(f"Validation data shape: {X_val_data.shape}")

# 5. Improved LSTM Model
def build_improved_lstm_model():
    # Extract text features and additional features
    text_input_dim = MAX_SEQUENCE_LENGTH
    additional_features_dim = train_features.shape[1]

    # Set hyperparameters
    embedding_dim = 200  # Increased from 128

    # Create model with both text input and additional features
    model = Sequential()

    # Embedding layer for text
    model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=text_input_dim))
    model.add(SpatialDropout1D(0.15))  # Reduced from 0.2

    # Bidirectional LSTM with return sequences
    model.add(Bidirectional(LSTM(128, dropout=0.15, recurrent_dropout=0.15, return_sequences=True)))

    # Global pooling to capture important features
    model.add(GlobalMaxPooling1D())

    # Dense layers
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.3))  # Reduced from 0.5
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, activation='sigmoid'))

    # Compile with reduced learning rate and accuracy only
    model.compile(
        loss='binary_crossentropy',
        optimizer=Adam(learning_rate=0.0005),  # Reduced from 0.001
        metrics=['accuracy']  # 只保留准確率指標
    )

    return model

# 6. Improved GRU Model
def build_improved_gru_model():
    # Extract text features and additional features
    text_input_dim = MAX_SEQUENCE_LENGTH
    additional_features_dim = train_features.shape[1]

    # Set hyperparameters
    embedding_dim = 200  # Increased from 128

    # Create model with both text input and additional features
    model = Sequential()

    # Embedding layer for text
    model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=text_input_dim))
    model.add(SpatialDropout1D(0.15))  # Reduced from 0.2

    # Bidirectional GRU with return sequences
    model.add(Bidirectional(GRU(128, dropout=0.15, recurrent_dropout=0.15, return_sequences=True)))

    # Global pooling to capture important features
    model.add(GlobalMaxPooling1D())

    # Dense layers
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.3))  # Reduced from 0.5
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, activation='sigmoid'))

    # Compile with reduced learning rate and accuracy only
    model.compile(
        loss='binary_crossentropy',
        optimizer=Adam(learning_rate=0.0005),  # Reduced from 0.001
        metrics=['accuracy']  # 只保留准確率指標
    )

    return model

# 7. Training and Evaluation with Improved Callbacks
batch_size = 64
epochs = 5

# Improved callbacks
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=7,  # Increased from 3
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=3,
    min_lr=0.00001,
    verbose=1
)

# Helper function to get memory usage
def get_memory_usage():
    return psutil.Process(os.getpid()).memory_info().rss / 1024 / 1024  # in MB

# Improved training function to measure time and memory
def train_model(model, model_name):
    print(f"\nTraining {model_name} model...")

    # Measure memory before training
    memory_before = get_memory_usage()
    print(f"Memory usage before training: {memory_before:.2f} MB")

    # Measure training time
    start_time = time.time()

    # Create F1 metrics callback
    f1_metrics = F1Metrics(validation_data=(X_val_data, y_val_data))

    # Train the model with class weights and improved callbacks
    history = model.fit(
        X_train_data, y_train_data,
        validation_data=(X_val_data, y_val_data),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stopping, reduce_lr, f1_metrics],
        class_weight=class_weight_dict,  # Add class weights
        verbose=1
    )

    # Calculate training time
    training_time = time.time() - start_time
    print(f"{model_name} training time: {training_time:.2f} seconds")

    # Measure memory after training
    memory_after = get_memory_usage()
    memory_usage = memory_after - memory_before
    print(f"Memory usage after training: {memory_after:.2f} MB")
    print(f"Memory increase during training: {memory_usage:.2f} MB")

    # Evaluate on validation set
    val_loss, val_acc = model.evaluate(X_val_data, y_val_data, verbose=0)
    print(f"{model_name} validation accuracy: {val_acc:.4f}")

    # Predictions on validation set
    y_pred = (model.predict(X_val_data) > 0.5).astype(int)

    # Calculate F1 score
    val_f1 = f1_score(y_val_data, y_pred)
    print(f"{model_name} validation F1 score: {val_f1:.4f}")

    # Print classification report
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_val_data, y_pred))

    # Calculate confusion matrix
    cm = confusion_matrix(y_val_data, y_pred)

    # Plot training history - 移除 precision 和 recall 相關的繪圖
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(f'{model_name} Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(f'{model_name} Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.tight_layout()
    plt.savefig(f'{model_name}_training_history.png')
    plt.show()

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'{model_name} Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.savefig(f'{model_name}_confusion_matrix.png')
    plt.show()

    return {
        'model': model,
        'history': history,
        'training_time': training_time,
        'memory_usage': memory_usage,
        'val_accuracy': val_acc,
        'val_f1': val_f1
    }

# Train improved LSTM model
lstm_model = build_improved_lstm_model()

# 在調用 summary() 前先強制構建模型以避免 "unbuilt" 問題
dummy_data = np.zeros((1, X_train_data.shape[1]))
_ = lstm_model(dummy_data)
lstm_model.summary()

lstm_results = train_model(lstm_model, "LSTM")

# Train improved GRU model
gru_model = build_improved_gru_model()

# 在調用 summary() 前先強制構建模型以避免 "unbuilt" 問題
_ = gru_model(dummy_data)
gru_model.summary()

gru_results = train_model(gru_model, "GRU")

# 8. Model Comparison with More Metrics
print("\n===== Model Comparison =====")
print(f"LSTM training time: {lstm_results['training_time']:.2f} seconds")
print(f"GRU training time: {gru_results['training_time']:.2f} seconds")
print(f"LSTM memory usage: {lstm_results['memory_usage']:.2f} MB")
print(f"GRU memory usage: {gru_results['memory_usage']:.2f} MB")
print(f"LSTM validation accuracy: {lstm_results['val_accuracy']:.4f}")
print(f"GRU validation accuracy: {gru_results['val_accuracy']:.4f}")
print(f"LSTM validation F1 score: {lstm_results['val_f1']:.4f}")
print(f"GRU validation F1 score: {gru_results['val_f1']:.4f}")

# Visualize improved comparison (移除 precision 和 recall 相關的比較)
metrics = ['Training Time (s)', 'Memory Usage (MB)', 'Accuracy', 'F1 Score']
lstm_values = [
    lstm_results['training_time'],
    lstm_results['memory_usage'],
    lstm_results['val_accuracy'],
    lstm_results['val_f1']
]
gru_values = [
    gru_results['training_time'],
    gru_results['memory_usage'],
    gru_results['val_accuracy'],
    gru_results['val_f1']
]

# Create a comparative bar chart
plt.figure(figsize=(15, 8))
x = np.arange(len(metrics))
width = 0.35

plt.bar(x - width/2, lstm_values, width, label='LSTM')
plt.bar(x + width/2, gru_values, width, label='GRU')

plt.title('LSTM vs GRU Performance Comparison')
plt.xticks(x, metrics)
plt.ylabel('Value')
plt.legend()

plt.tight_layout()
plt.savefig('model_comparison.png')
plt.show()

# 9. Making Predictions on Test Data
def generate_predictions(model, model_name):
    print(f"\nGenerating predictions with {model_name} model...")
    predictions = (model.predict(X_test) > 0.5).astype(int)
    return predictions.flatten()

# Generate predictions
lstm_predictions = generate_predictions(lstm_model, "LSTM")
gru_predictions = generate_predictions(gru_model, "GRU")

# 10. Creating CSV output
output_df = pd.DataFrame({
    'id': test_df['id'],
    'LSTM': lstm_predictions,
    'GRU': gru_predictions
})

print("\nPrediction output (first few rows):")
print(output_df.head())

# Save to CSV
output_df.to_csv('predictions.csv', index=False)
print("Predictions saved to 'predictions.csv'")

# 11. Creating ensemble predictions
# Create an ensemble prediction (average of both models)
ensemble_predictions = ((lstm_predictions + gru_predictions) > 0.5).astype(int)
output_df['Ensemble'] = ensemble_predictions

# Save ensemble predictions
output_df.to_csv('ensemble_predictions.csv', index=False)
print("Ensemble predictions saved to 'ensemble_predictions.csv'")

# 12. Summary with improved metrics (移除 precision 和 recall 相關的摘要)
print("\n===== Summary =====")
print(f"LSTM Model:")
print(f"- Training time: {lstm_results['training_time']:.2f} seconds")
print(f"- Memory usage: {lstm_results['memory_usage']:.2f} MB")
print(f"- Validation accuracy: {lstm_results['val_accuracy']:.4f}")
print(f"- Validation F1 score: {lstm_results['val_f1']:.4f}")

print(f"\nGRU Model:")
print(f"- Training time: {gru_results['training_time']:.2f} seconds")
print(f"- Memory usage: {gru_results['memory_usage']:.2f} MB")
print(f"- Validation accuracy: {gru_results['val_accuracy']:.4f}")
print(f"- Validation F1 score: {gru_results['val_f1']:.4f}")

print("\nModel Differences:")
time_diff = gru_results['training_time'] - lstm_results['training_time']
memory_diff = gru_results['memory_usage'] - lstm_results['memory_usage']
accuracy_diff = gru_results['val_accuracy'] - lstm_results['val_accuracy']
f1_diff = gru_results['val_f1'] - lstm_results['val_f1']

print(f"- Training time: GRU is {abs(time_diff):.2f} seconds {'faster' if time_diff < 0 else 'slower'} than LSTM")
print(f"- Memory usage: GRU uses {abs(memory_diff):.2f} MB {'less' if memory_diff < 0 else 'more'} memory than LSTM")
print(f"- Validation accuracy: GRU is {abs(accuracy_diff)*100:.2f}% {'better' if accuracy_diff > 0 else 'worse'} than LSTM")
print(f"- Validation F1 score: GRU is {abs(f1_diff)*100:.2f}% {'better' if f1_diff > 0 else 'worse'} than LSTM")

print("\nConclusion:")
if gru_results['val_f1'] > lstm_results['val_f1']:
    better_model = "GRU"
    better_metric = gru_results['val_f1']
elif lstm_results['val_f1'] > gru_results['val_f1']:
    better_model = "LSTM"
    better_metric = lstm_results['val_f1']
else:
    better_model = "Both models performed equally"
    better_metric = lstm_results['val_f1']

print(f"The {better_model} model performed better with an F1 score of {better_metric:.4f}.")
print("An ensemble of both models was also created, which may provide more robust predictions.")
```
